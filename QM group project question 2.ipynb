# Safety as a reason leading to differences in sexual offences across London boroughs

```python
#Import of the packages, data and creation of the dataframe for sexual offences 
import plotly.express as px
import pandas as pd
from scipy.stats import pearsonr
path = '/Users/jutti/Documents/QM project/met-police-recorded-offences-rates-borough.xlsx'

xls = pd.ExcelFile(path)

df = pd.read_excel(xls,sheet_name='Recorded Offences')

df
```

```python
#Cleaning the dataframe

i = 0
while(i<182):
    
    df.rename(columns={df.columns[i]: df.loc[0][i]}, inplace = True)
    i=i+1
    
df = df.loc[2:33]

df
```

```python
#We store the data for all indicators in seperate lists for further analysis. 

number_of_sexual_offences = []    
all_recorded_offences = []
violence_against_the_person = []
robbery = []
burglary = []
theft_and_handling = []
fraud_and_forgery = []
criminal_damage = []
drugs = []
borough =[]

row =2
column = 0
while(row<34):
     
    all_recorded_offences.append(df.loc[row][column+2])
    violence_against_the_person.append(df.loc[row][column+20])
    number_of_sexual_offences.append(df.loc[row][column+38])
    robbery.append(df.loc[row][column+56])
    burglary.append(df.loc[row][column+74])
    theft_and_handling.append(df.loc[row][column+92])
    fraud_and_forgery.append(df.loc[row][column+110])
    criminal_damage.append(df.loc[row][column+128])
    drugs.append(df.loc[row][column+146])
    borough.append(df.loc[row][1])
    column += 1
    if(column == 18):
        column = 0
        row = row+1


#Creation of final dataframe to plot the graphs
final_df= pd.DataFrame()

final_df['All recorded crime'] = all_recorded_offences

final_df['Violence against the person']= violence_against_the_person

final_df['Sexual assaults']=number_of_sexual_offences

final_df['Robberies']=robbery

final_df['Burglaries']=burglary

final_df['Theft and handling']=theft_and_handling

final_df['Fraud and forgery']=fraud_and_forgery

final_df['Criminal damage']=criminal_damage

final_df['Drugs']=drugs

final_df['Boroughs']=borough




#creation of first scatter plot 

fig = px.scatter(final_df,x='All recorded crime',y='Sexual assaults',color='Boroughs',
                 title = "Relationship between overall offences level and sexual offences in London boroughs",
                 trendline="ols",trendline_scope="overall")

fig.show()
fig.write_html('/Users/jutti/Documents/QM2SRQ1-2/Visualisations/Relationship between overall offences level and sexual offences in London boroughs.html')
```

```python
#We perform pearson correlation because we are looking for a linear relationship between the two datasets.
#We will also get the p-value from the pearsonr method

correlation_1 = pearsonr(all_recorded_offences,number_of_sexual_offences)

correlation_2 = pearsonr(violence_against_the_person,number_of_sexual_offences)

correlation_3 = pearsonr(robbery,number_of_sexual_offences)

correlation_4 = pearsonr(burglary,number_of_sexual_offences)

correlation_5 = pearsonr(theft_and_handling,number_of_sexual_offences)

correlation_6 = pearsonr(fraud_and_forgery,number_of_sexual_offences)

correlation_7 = pearsonr(criminal_damage,number_of_sexual_offences)

correlation_8 = pearsonr(drugs,number_of_sexual_offences)

print(correlation_1,correlation_2,correlation_3,correlation_4,
      correlation_5,correlation_6,correlation_7,correlation_8)
```

# Socio-economic reasons as reasons leading to differences in sexual offences across London boroughs

```python
#First we look at the relationship between income level of a borough and prevalence of secual offences


#Import of the packages and data for the number of sexual offences per borough and creation of dataframe
import plotly.express as px
import pandas as pd
from scipy.stats import pearsonr

path = '/Users/jutti/Documents/QM project/met-police-recorded-offences-rates-borough.xlsx'

xls = pd.ExcelFile(path)

df = pd.read_excel(xls,sheet_name='Recorded Offences')

df = df.loc[0:33][df.columns[40:56]]
df

```

```python
#We rename the columns as done before above and focus on sexual offences from 2002 to 2017


i = 0
while(i<16):
    
    df.rename(columns={df.columns[i]: df.loc[0][i]}, inplace = True)
    i=i+1
    

df = df.loc[2:33]

df

#storage of sexual offences per borough in list for 2003-2017

number_of_sexual_offences = []    
i =2
column = 0
while(i<34):
     
    number_of_sexual_offences.append(df.loc[i][column])
    column +=1
    if(column == 16):
        column = 0
        i = i+1
```

```python
#Import of the income data per borough and creation of dataframe

path2 = '/Users/jutti/Documents/QM project/earnings-residence-borough.xlsx'

xlsx = pd.ExcelFile(path2)

df2 = pd.read_excel(xlsx,sheet_name='Total, weekly')

df2
```

```python
#reduction to rows and columns of interest and renaming of columns
df2 = df2.loc[3:34]
d = 3

while(d<20):
    df2.drop(df2.columns[d],axis =1, inplace = True)
    d = d+1

df2.pop('Code')

df2.drop(df2.columns[17:22],axis =1, inplace = True) 

df2
```

```python
#storage of weekly income per borough in list for 2003-2017

weekly_earnings =[]
boroughs =[]
row =3
column2 = 1
while(row<35):
     
    weekly_earnings.append(df2.loc[row][df2.columns[column2]])
    boroughs.append(df2.loc[row][df2.columns[0]])
    column2 +=1
    if(column2 == 17):
        column2 = 1
        row = row+1



#conversion to yearly earnings

yearly_earnings = [item*52 for item in weekly_earnings]


#Creation of final datadrame with which we will construct the plot

final_df= pd.DataFrame()

final_df['yearly earnings'] = yearly_earnings

final_df['sexual assaults']= number_of_sexual_offences

final_df['boroughs']=boroughs

#final_df['boroughs']=boroughs

#creation of scatter plot 

fig = px.scatter(final_df,x='yearly earnings',y='sexual assaults',color='boroughs',
                 title = "Relationship between income level and sexual offences in London boroughs",
                trendline="ols", trendline_scope="overall")

fig.show()
fig.write_html('/Users/jutti/Documents/QM2SRQ1-2/Visualisations/Relationship between income level and sexual offences in London boroughs.html')
```

```python
#calculation of the correlation between income and sexual assaults

correlation = pearsonr(weekly_earnings,number_of_sexual_offences)

print(correlation)
```
